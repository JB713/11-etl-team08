# Read and transform video games sales data

### Import libraries

```python
import pandas as pd
```

### Read data and tranform

```python
VGSales_file = "source_data/vgsales.csv"
VGSales_df = pd.read_csv(VGSales_file)
VGSales_df.head()
```

```python
VGNew=VGSales_df.drop(columns=['Rank'])
VGNew.head()
```

```python
indexNames = VGNew[ (VGNew['Year'] <= 1996)].index
VGNew.drop(indexNames , inplace=True)
```

```python
VGNew.head()
```

```python
indexNames = VGNew[ (VGNew['Year'] >= 2017)].index
VGNew.drop(indexNames , inplace=True)
```

```python
VGSF=VGNew.drop(columns=['Genre','EU_Sales','JP_Sales','Other_Sales','Global_Sales'])
VGSF.head()
```

```python
VGSF.groupby(by=["Publisher"]).sum()
```

```python
VGSF.groupby(by=["Year"]).count()
```

```python
VGSF.groupby(by=["Platform"]).count()
```

```python
VGSF.loc[VGSF['Platform']=="3DS",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="DS",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']== "GB",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="GBA",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="GC",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="N64",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="PC",'Company'] = 'Microsoft'
VGSF.loc[VGSF['Platform']=="PS",'Company'] = 'SONY'
VGSF.loc[VGSF['Platform']=="PS2",'Company'] = 'SONY'
VGSF.loc[VGSF['Platform']=="PS3",'Company'] = 'SONY'
VGSF.loc[VGSF['Platform']=="PS4",'Company'] = 'SONY'
VGSF.loc[VGSF['Platform']=="PSP",'Company'] = 'SONY'
VGSF.loc[VGSF['Platform']=="PSV",'Company'] = 'SONY'
VGSF.loc[VGSF['Platform']=="Wii",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="WiiU",'Company'] = 'Nintendo'
VGSF.loc[VGSF['Platform']=="X360",'Company'] = 'Microsoft'
VGSF.loc[VGSF['Platform']=="XB",'Company'] = 'Microsoft'
VGSF.loc[VGSF['Platform']=="XOne",'Company'] = 'Microsoft'
VGSF['Company'].fillna('Other', inplace=True)
```

```python
VGSF.head()
```

```python
VGSF.groupby(['Year','Publisher']).count()
```

### Create new clean CSV file

```python
VGSF.to_csv('VGsales_Clean.csv')
```
